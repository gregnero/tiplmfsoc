```python
def create_scatter_plot(x, y, xlabel, ylabel, title):
    # Private styling variables (for internal consistency)
    _font_size_labels = 14
    _font_size_ticks = 12
    _font_size_title = 13
    _marker_size = 50

    # Create the figure and axis
    fig, ax = plt.subplots(figsize=(5, 5), dpi=300)  # Square figure size

    # Scatter plot
    ax.scatter(x, y, s=_marker_size, edgecolors='black', linewidths=0.5)

    ax.set_yscale('log')

    # Axis labels and title
    ax.set_xlabel(xlabel, fontsize = _font_size_labels)
    ax.set_ylabel(ylabel, fontsize = _font_size_labels)
    ax.set_title(title, fontsize = _font_size_title)

    # Tick parameters
    ax.tick_params(axis='both', which='major', labelsize=_font_size_ticks)

    # Optional: grid for better readability
    ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

    # Tight layout for clean spacing
    fig.tight_layout()

create_scatter_plot(received_powers_nW, uncoded_BER_values, "Rx Power [nW]", "Uncoded BER", "(b) Uncoded BER vs. Rx Power over 750m")
```
